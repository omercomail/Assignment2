```python

df1 = pd.read_csv("Merged_cab_data.csv")

# Aggregate data
df1 = df1.groupby(["Company", "City"], as_index=False).agg({"Travelled_Km": "sum"})
# Create list to sort by
sorter = (
    df1.groupby(["Company"], as_index=False)
    .agg({"Travelled_Km": "sum"})
    .sort_values("Travelled_Km")["Company"]
    .values
)

# Pivot, reindex, and plot
df1 = df1.pivot(index="Company", columns="City", values="Travelled_Km")
df1 = df1.reindex(sorter)
df1.plot.bar(position=1, width=0.3,stacked=True)
```

```python

df1 = pd.read_csv("Merged_cab_data.csv")

# Aggregate data
df1 = df1.groupby(["Company", "Customer_ID"], as_index=False).agg({"Income": "sum"})
# Create list to sort by
sorter = (
    df1.groupby(["Company"], as_index=False)
    .agg({"Income": "sum"})
    .sort_values("Income")["Company"]
    .values
)

# Pivot, reindex, and plot
df1 = df1.pivot(index="Company", columns="Customer_ID", values="Income")
df1 = df1.reindex(sorter)
df1.plot.bar(position=1, width=0.3,stacked=True)
```
